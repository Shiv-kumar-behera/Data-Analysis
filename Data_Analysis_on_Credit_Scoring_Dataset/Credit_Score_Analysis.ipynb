## Categorical Columns
- Occupation :- (The job or profession of the individual.)
- Credit_Mix :- (The mix of different types of credit a person has :- (e.g., credit cards, loans).)
- Payment_of_Min_Amount :- (Indicates if the person is making the minimum required payment on their credit.)
- Payment_Behaviour :- (Describes the pattern of payments the individual makes :- (e.g., on time, delayed).)
- Credit_Score :- (The credit score of the individual, which is the target variable to be predicted.)

## Numerical Columns
- Age :- (The age of the individual.)
- Annual_Income :- (The total yearly income of the person.)
- Num_Bank_Accounts :- (The number of bank accounts the individual has.)
- Num_Credit_Card :- (The number of credit cards the individual has.)
- Interest_Rate :- (The interest rate on a loan.)
- Num_of_Loan :- (The total number of loans a person has.)
- Delay_from_due_date :- ( The number of days a payment was delayed from its due date.)
- Num_of_Delayed_Payment :- (The total count of delayed payments.)
- Changed_Credit_Limit :- (This likely indicates whether the credit limit was changed :- (e.g., increased or decreased).)
- Num_Credit_Inquiries :- (The number of credit inquiries made on the person's credit report.)
- Outstanding_Debt :- (The total amount of money owed.)
- Credit_Utilization_Ratio :- (The percentage of a person's total available credit that they're currently using.)
- Total_EMI_per_month :- (The total amount of Equated Monthly Installment payments per month.)
- Amount_invested_monthly :- (The amount of money invested by the person on a monthly basis.)
- Monthly_Balance :- (The monthly balance of the individual's accounts.)
- Credit_History_Age_Months	:- (he length of a person's credit history in months.)


## Columns whoose data types needs to be changes for simplicity
- Annual_income (to int64)
- Changed_Credit_Limit (to int64)
- Num_Credit_Inquiries (to int64)
- Outstanding_Debt (to int64)
- Total_Emi (to int64)
- Amount_invested_monthly (to int64)
- Monthly_Balance (to int64)

```python
df["Annual_Income"] = df["Annual_Income"].astype("int")
df["Changed_Credit_Limit"] = df["Changed_Credit_Limit"].astype("int")
df["Num_Credit_Inquiries"] = df["Num_Credit_Inquiries"].astype("int")
df["Outstanding_Debt"] = df["Outstanding_Debt"].astype("int")
df["Total_EMI_per_month"] = df["Total_EMI_per_month"].astype("int")
df["Amount_invested_monthly"] = df["Amount_invested_monthly"].astype("int")
df["Monthly_Balance"] = df["Monthly_Balance"].astype("int")
```

# Occupation

# Age

```python
sns.histplot(df["Age"], kde=True, stat="density", kde_kws=dict(cut=3), edgecolor=(1, 1, 1, .4))
print(df["Age"].skew())
print(df["Age"].kurt())
```

- can be considered normal after some transformation

```python
sns.histplot(np.log(df["Age"]), kde=True, stat="density", kde_kws=dict(cut=3), edgecolor=(1, 1, 1, .4))
print(np.log(df["Age"]).skew())
print(np.log(df["Age"]).kurt())
```

# Annual Income

```python
q1 = df["Annual_Income"] > df["Annual_Income"].quantile(0.05)
q2 = df["Annual_Income"] < df["Annual_Income"].quantile(0.95)
x = df[q1 & q2]
sns.histplot(x["Annual_Income"], kde=True, stat="density", kde_kws=dict(cut=3), edgecolor=(1, 1, 1, .4))
print(x["Annual_Income"].skew())
print(x["Annual_Income"].kurt())
```
