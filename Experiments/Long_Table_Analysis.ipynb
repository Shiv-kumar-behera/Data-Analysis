```python
import pandas as pd
import numpy as np
```

```python
Persons = ["Shiv", "Ankur", "Ankit", "Rajeev", "Saswat"]
Users = Persons * 365 * 2
Groceries = np.int64(np.random.power(a=0.05, size=3650) * 10000)
Clothing = np.int64(np.random.power(a=0.01, size=3650) * 50000)
Housing = np.int64(np.random.power(a=0.01, size=3650) * 30000)
Education = np.int64(np.random.power(a=0.01, size=3650) * 50000)
Health = np.int64(np.random.power(a=0.01, size=3650) * 10000)
Travel = np.int64(np.random.power(a=0.01, size=3650) * 20000)
Entertainment = np.int64(np.random.power(a=0.6, size=3650) * 20000)
Gambling = np.int64(np.random.power(a=0.01, size=3650) * 100000)
Tax = np.int64(np.random.power(a=0.01, size=3650) * 10000)
Fines = np.int64(np.random.power(a=0.01, size=3650) * 5000)
dates = pd.date_range(start="2022-01-01", periods=365*2, freq="D")
df = pd.DataFrame({
    "Date": np.repeat(dates, 5),
    "Users": Users,
    "Groceries": Groceries,
    "Clothing": Clothing,
    "Housing": Housing,
    "Education": Education,
    "Health": Health,
    "Travel": Travel,
    "Entertainment": Entertainment,
    "Gambling": Gambling,
    "Tax": Tax,
    "Fines": Fines
})
```

```python
df.head(10)
```

```python
df["Yearly"] = df["Date"].dt.year
df["Half-Yearly"] = np.where(df["Date"].dt.month <= 6, 1, 2)
df["Half-Yearly"] = df["Yearly"].astype(str) + '-' + df["Half-Yearly"].astype(str)
df["Quaterly"] = df["Yearly"].astype(str) + '-' + df["Date"].dt.quarter.astype(str)
df["Monthly"] = df["Yearly"].astype(str) + '-' + df["Date"].dt.month.astype(str)
df["Weekly"] = df["Yearly"].astype(str) + '-' + df["Date"].dt.isocalendar().week.astype(str)
df.drop(columns="Date", inplace=True)
```

```python
x = df.groupby(["Users", "Yearly", "Half-Yearly", "Quaterly", "Monthly", "Weekly"]).sum()
x.to_csv("Data.csv")
x.sample(50)
```

```python
dic = {"Weekly": "W",
       "Monthly": "M",
       "Quaterly": "3M",
       "Half-Yearly": "6M",
       "Yearly": "Y"
}
data_frames = {}
for key, value in dic.items():
    data_frames[key] = df.groupby("Users")[["Groceries",
    "Clothing",
    "Housing",
    "Education",
    "Health",
    "Travel",
    "Entertainment",
    "Gambling",
    "Tax",
    "Fines"]].resample(value).sum().groupby("Users")
```

```python
data_frames["Weekly"].sum()
```

```python
data_weekly = {
    "Max": data_frames["Weekly"].max().T,
    "Min": data_frames["Weekly"].min().T,
    "Mean": data_frames["Weekly"].mean().T,
    "Median": data_frames["Weekly"].median().T
}
data1 = pd.concat(
    data_weekly,
    names=["Functions"]
)
data1
```

```python
data_monthly = {
    "Max": data_frames["Monthly"].max().T,
    "Min": data_frames["Monthly"].min().T,
    "Mean": data_frames["Monthly"].mean().T,
    "Median": data_frames["Monthly"].median().T
}
data2 = pd.concat(
    data_monthly,
    names=["Functions"]
)
data2
```

```python
data_quaterly = {
    "Max": data_frames["Quaterly"].max().T,
    "Min": data_frames["Quaterly"].min().T,
    "Mean": data_frames["Quaterly"].mean().T,
    "Median": data_frames["Quaterly"].median().T
}
data3 = pd.concat(
    data_quaterly,
    names=["Functions"]
)
data3
```

```python
data_half_yearly = {
    "Max": data_frames["Half-Yearly"].max().T,
    "Min": data_frames["Half-Yearly"].min().T,
    "Mean": data_frames["Half-Yearly"].mean().T,
    "Median": data_frames["Half-Yearly"].median().T
}
data4 = pd.concat(
    data_half_yearly,
    names=["Functions"]
)
data4
```

```python
data_yearly = {
    "Max": data_frames["Yearly"].max().T,
    "Min": data_frames["Yearly"].min().T,
    "Mean": data_frames["Yearly"].mean().T,
    "Median": data_frames["Yearly"].median().T
}
data5 = pd.concat(
    data_yearly,
    names=["Functions"]
)
data5
```

```python
data_merged = {
    "Weekly": data1,
    "Monthly": data2,
    "Quaterly": data3,
    "Half-Yearly": data4,
    "Yearly": data5
}
data = pd.concat(
    data_merged,
    names=["Aggr. Types"]
)
```
